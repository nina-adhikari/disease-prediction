```python
def get_keys_with_value_one(dictionary):
  """
  This function returns a list of keys in a dictionary whose value is 1.

  Args:
    dictionary: The dictionary to search.

  Returns:
    A list of keys in the dictionary whose value is 1.
  """

  keys_with_value_one = []
  for key, value in dictionary.items():
    if value == 1:
      keys_with_value_one.append(key)
  return keys_with_value_one
```

```python
from collections import defaultdict

for subset in SUBSETS:
    dfq = defaultdict(list)

    for (key, val) in get_keys_with_value_one(df[subset][features_with_1s].stack()):
        dfq[key].append(val)

    dfq = dict(dfq)
    df[subset]['Evidences'] = dfq
```

```python
def sex(letter):
    if letter == 'M':
        return 'male'
    elif letter == 'F':
        return 'female'
```

```python
def deal_with_lesion(row):
    text = ''
    if row['lesion_color'] not in ['NA', 'N', '0', 0, 'nowhere']:
        text += " of color " + row['lesion_color']
    #if row['lesion_pain_swollen'] not in ['NA', 'N', '0', 0, 'nowhere']:
        #text += " with a " + SYMPTOM_PHRASES['lesion_pain_swollen'] + " of " + str(row['lesion_pain_swollen'])
    text += " in my " + body_parts[row['lesion_location']] + " with a " + SYMPTOM_PHRASES['lesion_pain_intense'] + " of " + str(row['lesion_pain_intense'])
    return text

def deal_with_swelling(row):
    text = ''
    if row['swelling_location'] not in ['NA', 'N', '0', 0, 'nowhere']:
        text += " in my " + body_parts[row['swelling_location']]
    return text

def deal_with_pain(row):
    text = ''
    if row['pain_somewhere'] not in ['NA', 'N', '0', 0, 'nowhere']:
        text += " in my " + body_parts[row['pain_somewhere']]
    if row['pain_intensity'] not in ['NA', 'N', '0', 0, 'nowhere']:
        text += " with a " + SYMPTOM_PHRASES['pain_intensity'] + " of " + str(row['pain_intensity'])
    if row['pain_char'] not in ['NA', 'N', '0', 0, 'nowhere']:
        text += " which can be characterized as " + str(row['pain_char'])
    return text
```

```python
def get_text(row):
    pain_done = False
    lesion_done = False
    swelling_done = False
    text = ""
    text += "I am " + sex(row['SEX']) + " and " + str(row['AGE']) + " years old. "
    init_evi = row['INITIAL_EVIDENCE']
    text += "I have been experiencing " + SYMPTOM_PHRASES[init_evi]
    if init_evi == 'lesions':
        text += deal_with_lesion(row)
        lesion_done = True
    if init_evi == 'swelling':
        text += deal_with_swelling(row)
        swelling_done = True
    if init_evi == 'pain':
        text += deal_with_pain(row)
        pain_done = True
    text += '. '
    firstitem = True
    for symptom in features_with_1s:
        if row[symptom] == 1:
            if firstitem:
                text += "Additional symptoms include "
                firstitem = False
            else:
                text += ", "
            text += SYMPTOM_PHRASES[symptom]
            if symptom == 'lesions' and  lesion_done is False:
                text += deal_with_lesion(row)
            if symptom == 'swelling' and swelling_done is False:
                text += deal_with_swelling(row)
            if symptom == 'pain' and pain_done is False:
                text += deal_with_pain(row)
    text += ". "
    return text
```

```python
for subset in SUBSETS:
    df[subset]['Text'] = df[subset].apply(get_text, axis=1)
```
