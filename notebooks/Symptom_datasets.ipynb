```python
evidences['possible_values_en'] = [list() for n in range(len(evidences))]

for row in evidences.itertuples():
  vals = row.possible_values
  if len(row.value_meaning) == 0:
    continue
  for value in vals:
    row.possible_values_en.append(row.value_meaning[value]['en'])
```

```python
import ast

def dds_to_dicts(dataframe):
  dataframe['DIFFERENTIAL_DIAGNOSIS'] = [dict(ast.literal_eval(thing)) for thing in dataframe['DIFFERENTIAL_DIAGNOSIS'].values]

def evidences_to_lists(dataframe):
  dataframe['EVIDENCES'] = [ast.literal_eval(thing) for thing in dataframe['EVIDENCES'].values]
```

Translate to English. Still need to implement for Evidences:

```python
def translate_to_english(dataframe):
  dataframe['DIFFERENTIAL_DIAGNOSIS'] = [{conditions.loc[k]['cond-name-eng']: v.pop(k) for k in list(v.keys())} for v in dataframe['DIFFERENTIAL_DIAGNOSIS']]
  dataframe['PATHOLOGY'] = [conditions.loc[k]['cond-name-eng'] for k in dataframe['PATHOLOGY']]
  dataframe['INITIAL_EVIDENCE'] = [evidences.loc[k]['question_en'] for k in dataframe['INITIAL_EVIDENCE']]
```

```python
for dataframe in list(df.values()):
  dds_to_dicts(dataframe)
  evidences_to_lists(dataframe)
  translate_to_english(dataframe)
```
